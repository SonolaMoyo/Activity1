```python
#SONOLA MOYOSOLUWALORUN
#ID: 1479301970c1f000

#measuring regression performance
#MEAN ABSOLUTE ERROR MEAN ABSOLUTE ERROR

#firstly we normalise our dataset to a common scale using the min max scaler
from sklearn.preprocessing import MinMaxScaler
scalar = MinMaxScaler()

#we now drop the columns Date and Lights
df = df.drop(columns=['Date', 'Lights_wh'])

#normalise the whole dataset to a common scale
normalised_df = pd.DataFrame(scalar.fit_transform(df), columns=column_names2)

#rename the normalised df
normalised_df = normalised_df.rename(columns=column_names2)

features_df = normalised_df.drop(columns=['Appliances_wh'])
#extract appliance_target
appliance_target = normalised_df['Appliances_wh']

```

```python
#now we split our dataset into training and testing Dataset. Recall that we had already segmented the featured
#and target variables
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(features_df, appliance_target, test_size=0.3, random_state=42)
```

```python
#import linear regression function
from sklearn.linear_model import LinearRegression
linear_model = LinearRegression()
#fit the model into the training dataset
linear_model.fit(x_train, y_train)

#obtain predictions
predicted_values = linear_model.predict(x_test)

#now obtain MAE
from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(y_test, predicted_values)
print('The Mean absolute Error',round(mae, 2)) 
```

```python
#Residual Sum of Squares (RSS)

rss = np.sum(np.square(y_test - predicted_values))
print('Residual Sum of Squares:', round(rss, 2))
```

```python
#Root Mean Square Errors (RMSE)

from sklearn.metrics import mean_squared_error
rmse = np.sqrt(mean_squared_error(y_test, predicted_values))
print('Root Mean Square Error: ', round(rmse, 3))
```

```python
#R-Squared

from sklearn.metrics import r2_score
r2_score = r2_score(y_test, predicted_values)
print('R-Squared value gives: ',round(r2_score, 2))
```

```python
#ridge regression
from sklearn.linear_model import Ridge
rigde_reg = Ridge(alpha=0.4)
rigde_reg.fit(x_train, y_train)
```

```python
#lasso regression
from sklearn.linear_model import Lasso
lasso_reg = Lasso(alpha=0.001)
lasso_reg.fit(x_train, y_train)
```

```python
#Comparing the effect of regularisation
#this fuuction returns the weight of every feature
def get_weights_df(model, feat, col_name):
    #this function returns the weight of every feature
    weights = pd.Series(model.coef_, feat.columns).sort_values()
    weights_df = pd.DataFrame(weights).reset_index()
    weights_df.columns = ['Features', col_name]
    weights_df[col_name].round(3)
    return weights_df
```

```python
#gets the model individually
linear_model_weights = get_weights_df(linear_model, x_train, 'Linear_Model_Weight')
ridge_weights_df = get_weights_df(rigde_reg, x_train, 'Rigde_Weight')
lasso_weights_df = get_weights_df(lasso_reg, x_train, 'Lasso_weight')

final_weights = pd.merge(linear_model_weights, ridge_weights_df, on='Features')
final_weights = pd.merge(final_weights, lasso_weights_df, on='Features')

final_weights

#SONOLA MOYOSOLUWALORUN
#ID: 1479301970c1f000
```
